```python
import re
```

I first defined my helper functions using CoPilot

```python
"""Used CoPilot with the following command"""
# Write a regular expression that extracts the numbers, decimals, and floats between sequences of letters.
"""
I also used CoPilot to generate additional comments with examples so that CoPilot generates 
a function with the behavior that I want. I also modified the last two comments here,
to account for decimal stoichiometric coefficients
"""
# The function should return a dictionary of all sequences found, in the order that they are encountered.
# For example, given the string "a1b2c3", the function should return {"a": 1, "b": 2, "c": 3}.
# Given the string "a10b20c30", the function should return {"a": 10, "b": 20, "c": 30}.
# Given the string "abbh10b20c30", the function should return {"abbh": 10, "b": 20, "c": 30}.
# Given the string "a0.1b2", the function should return {"a": 0.1, "b": 2}
# Given the string "a0.1b2.3", the function should return {"a": 0.1, "b": 2.3}
def extract_numbers_from_string(string):
    return {key: float(value) for key, value in re.findall(r'([a-zA-Z]+)(\d+\.?\d*)', string)}

# Write a function that checks if the values in a dictionary sum to 1, and if they do not, it normalizes the values so that they do.
# The function should return a dictionary of all sequences found, in the order that they are encountered.
# For example, given the dictionary {"a": 1, "b": 2, "c": 3}, the function should return {"a": 0.1, "b": 0.2, "c": 0.3}.
# Given the dictionary {"a": 10, "b": 20, "c": 30}, the function should return {"a": 0.1, "b": 0.2, "c": 0.3}.
# Given the dictionary {"abbh": 10, "b": 20, "c": 30}, the function should return {"abbh": 0.1, "b": 0.2, "c": 0.3}.
# Given the dictionary {"a": 0.1, "b": 2}, the function should return {"a": 0.1, "b": 0.2}.
# Given the dictionary {"a": 0.1, "b": 2.3}, the function should return {"a": 0.1, "b": 0.2}.
# Given the dictionary {"a": 0.1, "b": 2.3, "c": 0.1}, the function should return {"a": 0.1, "b": 0.2, "c": 0.1}.
def normalize_dictionary(dictionary):
    sum = 0
    for key, value in dictionary.items():
        sum += value
    if sum == 1:
        return dictionary
    else:
        return {key: value/sum for key, value in dictionary.items()}

# Write a function that checks if the values in a dictionary contain more than two decimal places, and if they do, it rounds the values to two decimal places. Also if the value has all zeros after the decimal point, it should be rounded to the nearest integer.
# The function should return a dictionary of all sequences found, in the order that they are encountered.
# For example, given the dictionary {"a": 1, "b": 2, "c": 3}, the function should return {"a": 1, "b": 2, "c": 3}.
# Given the dictionary {"a": 10, "b": 20, "c": 30}, the function should return {"a": 10, "b": 20, "c": 30}.
# Given the dictionary {"abbh": 10, "b": 20, "c": 30}, the function should return {"abbh": 10, "b": 20, "c": 30}.
# Given the dictionary {"a": 0.1, "b": 2}, the function should return {"a": 0.1, "b": 2}.
# Given the dictionary {"a": 0.1, "b": 2.3}, the function should return {"a": 0.1, "b": 2.3}.
# Given the dictionary {"a": 0.1, "b": 2.3, "c": 0.1}, the function should return {"a": 0.1, "b": 2.3, "c": 0.1}.
# Given the dictionary {"a": 0.123, "b": 2.345, "c": 0.123}, the function should return {"a": 0.12, "b": 2.35, "c": 0.12}.
def round_dictionary(dictionary):
    return {key: round(value, 2) for key, value in dictionary.items()}

# Write a function that concatenates the keys and values of a dictionary into a string.
def concatenate_dictionary(dictionary):
    return ''.join([key + str(value) for key, value in dictionary.items()])
```

Using these helper functions, I manually created a function that normalizes the alloy name, *normalize_alloy*

```python
def normalize_alloy(alloy):
    dictionary = normalize_dictionary(extract_numbers_from_string(alloy))
    return concatenate_dictionary(round_dictionary(dictionary))
```

```python
print(normalize_alloy("A1B2"))
print(normalize_alloy("Ti6Al4V1"))
print(normalize_alloy("Ti0.6Al0.4V1"))
```

Conducting correlation analysis between features:

```python
## Used CoPilot
# Create function that takes in a dataframe and conducts a correlation analysis between 
# all columns. If the correlation between two columnds is greater than abs(0.8), 
# it should drop one of the columns. 
# The function should return a new dataframe with the columns that have a correlation less than abs(0.8).

""" 
I had to manually import numpy as np. I also manually change the function so that
it returns a correlation matrix, and to_drop (so that I can verify if the function works correctly)
"""
import numpy as np
def correlation_analysis(df):
    corr = df.corr().abs()
    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]
    return df.drop(df[to_drop], axis=1), corr, to_drop
```

```python
low_corr_featurized_df, correlation_matrix, features_to_drop = correlation_analysis(featurized_df)
```

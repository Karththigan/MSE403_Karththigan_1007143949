### <span style="color: red"> Task 2:</span> Who has the most runner-up (i.e., second place) finishes in the histroy of NHL? Create a table with the teams and their runner-up finishes. The table must contain the the following columns:

-Team

-Total number of runner-up finishes

-Years [list of the years they finished as a runner-up]

You have to use Jupyter for the practice and produce a notebook with good readability.

### <span style="color: blue"> Solution:</span>

```python
import pandas as pd
```

```python
# Import data
file = "Stanely_cup_winners_by_year.xlsx"
df = pd.read_excel(file)
```

```python
# Visuzlize structure of data
df.head()
```

```python
# Create new dataframe to store how many times each team was a runner up
df2 = pd.DataFrame()
```

```python
# Get all teams who were runner ups

# Learnt .unique() from
# https://stackoverflow.com/questions/32072076/find-the-unique-values-in-a-column-and-then-sort-them
df2['Team'] = df['Runner up'].unique()
```

```python
df2.head()
```

```python
runner_up_counts = pd.DataFrame(df.groupby(['Runner up']).count()['Team'])
```

```python
runner_up_counts.head()
```

```python
# Merged df2 with runner up counts 
# Learnt about pd.merge() from:
# https://towardsdatascience.com/all-the-pandas-merge-you-should-know-for-combining-datasets-526b9ecaf184
df2 = pd.merge(df2, runner_up_counts, left_on='Team', right_on='Runner up')
```

```python
df2.head()
```

```python
# Renaming column titles of runner-up history dataframe
df2.columns = ["Team", "Total number of runner-up finishes"]
```

```python
# List years each team was a runner-up by grouping by runner up
"""
Learnt this method of indexing what .groupby() returns by year and applying .apply() to 
# apply the list() function row by row from
# https://stackoverflow.com/questions/22219004/how-to-group-dataframe-rows-into-list-in-pandas-groupby
"""
years_as_runner_up = df.groupby('Runner up')['Year'].apply(list)
years_as_runner_up = pd.DataFrame(years_as_runner_up)

# Adding index column 
# .reset_index() method from 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html
years_as_runner_up = years_as_runner_up.reset_index()
```

```python
years_as_runner_up.head()
```

```python
# Merge df2 and years_as_runner_up to attach list of years each team was a runner up 
# to the team name
team_years_runner_up = pd.merge(df2, years_as_runner_up, left_on="Team", right_on='Runner up')
```

```python
df2['Years'] = team_years_runner_up['Year']
```

```python
df2
```

### <span style="color: blue"> Solution:</span>

```python
import matminer
from matminer.datasets import load_dataset
#matminer.datasets.get_available_datasets() #you can check what datasets are available with this line
```

```python
steel_data = load_dataset('steel_strength')
```

```python
steel_data.head()
```

Checking for NaN and duplicate rows:

```python
# Checking for NaNs per column
# https://stackoverflow.com/questions/56261130/count-all-nans-in-a-pandas-dataframe
steel_data.isna().sum(axis=0)
```

There are 9 formulae with NaN elongation values. These rows will be discarded

```python
steel_data = steel_data.dropna()
```

```python
# Checking for duplicate rows
# Using same methodology as checking for NaN but used .duplicated()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.duplicated.html
steel_data.duplicated().sum()
```

There are no repeated alloys in the data set.

```python
from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.composition import ElementFraction
from tqdm.autonotebook import tqdm
```

```python
# Get list of elements in each formula
steel_data_featurized = StrToComposition().featurize_dataframe(steel_data, "formula")
```

```python
# Get element fraction of each alloy in steel_data
steel_data_featurized = ElementFraction().featurize_dataframe(steel_data_featurized, "composition")
```

```python
copy_steel_data_featurized = steel_data_featurized
```

```python
steel_data_featurized.head()
```

```python
# Determining which columns are the featurized elemet fractions by finding
# the index of H in the columns
list(steel_data_featurized.columns).index('H')
```

Performing a correlation analysis between yield strength and element fraction:

```python
# Store correlation coefficients in dictionary for reference
ys_element_correlations = {}

for feature in steel_data_featurized.columns[18:]:
    correlation_coeff = steel_data_featurized['yield strength'].corr(steel_data_featurized[feature])
    ys_element_correlations[feature] = correlation_coeff
    
    if correlation_coeff > abs(0.8):
        steel_data_featurized = steel_data_featurized.drop(feature, axis=1)
```

```python
steel_data_featurized.head()
```

```python
[print(x) for x in ys_element_correlations.values() if x > abs(0.8)]
```

No features appear to have a correlation coefficient greater than abs(0.8) and so, no featurized element fractions were dropped.

Reattempting correlation analysis but with the given element fractions from the original 'steel_strength' dataset:

```python
ys_builtin_element_correlations = {}

for feature in copy_steel_data_featurized.columns[1:14]:
    correlation_coeff = copy_steel_data_featurized['yield strength'].corr(copy_steel_data_featurized[feature])
    ys_builtin_element_correlations[feature] = correlation_coeff
    
    if correlation_coeff > abs(0.8):
        steel_data_featurized = copy_steel_data_featurized.drop(feature, axis=1)
```

```python
copy_steel_data_featurized.head()
```

Again, no element fractions (rrom the original 'steel_strength' dataset from matminer) appear to have a correlation coefficient greater than abs(0.8) and so, no featurized element fractions were dropped.
